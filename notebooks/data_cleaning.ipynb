Cleaning and watching through anime dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
anime = pd.read_csv('../data/anime_filtered.csv').set_index('anime_id')
users = pd.read_csv('../data/users_filtered.csv').set_index('user_id')
```

```python
users.head()
```

```python
anime.head()
```

```python
print(anime.info())
print(users.info())
```

```python
anime.drop('title_english', axis=1, inplace=True)
anime.drop('title_synonyms', axis=1, inplace=True)
anime.drop('title_japanese', axis=1, inplace=True)
```

```python
def write_value_counts_to_file(df, name:str):
    values = pd.DataFrame()
    for column in df.columns:
        if df[column].nunique() > 100:
            continue
        s1 = pd.Series(index=[column], data=[0])
        values = pd.concat([values, s1])
        values = pd.concat([values, df[column].value_counts()])
    values.to_csv(f'../data/{name}_value_counts.csv')
            

write_value_counts_to_file(anime, 'anime')
write_value_counts_to_file(users, 'users')
        
```

Categorical columns:
type, source, status, airing, rating

```python
cols_to_drop = ['image_url', 'opening_theme', 'ending_theme', 'broadcast', 'licensor', 'background', 'related', 'producer']
anime.drop(cols_to_drop, axis=1, inplace=True)
```

```python
anime.head()
```

```python
anime['genre'].fillna('None', inplace=True)
```

```python
anime['episodes'].describe()
```

```python
anime['anime_duration'] = pd.cut(anime['episodes'], bins=[0, 12, 51, 100, np.inf], labels=['short', 'medium', 'long', 'very_long'], right=True)
```

```python
anime['anime_duration'].value_counts()
```

```python
df_encoded = anime['genre'].str.get_dummies(sep=',').add_prefix('genre_')
df_encoded.head()
anime = pd.concat([anime, df_encoded], axis=1)
anime.drop('genre', axis=1, inplace=True)
```

```python
anime.drop('episodes', axis=1, inplace=True)
```

```python
anime['duration'] = anime['duration'].str.extract('(\d+)')
anime['duration'] = anime['duration'].astype(float)
anime['duration'].describe()
```

```python
anime['duration'] = pd.cut(anime['duration'], bins=[0, 20, 35, np.inf], labels=['short', 'medium', 'long'], right=True)
anime['duration'].value_counts()
```

```python
anime['premiered_year'] = anime['aired'].str.extract('(\d{4})').astype(float)
anime['premiered_year'].describe()
```

```python
anime.drop(['premiered', 'status', 'airing', 'aired_string', 'aired'], axis=1, inplace=True, errors='ignore')
anime.head()
```

```python
sns.heatmap(anime.isnull(), cbar=False)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
numeric_columns = ['score', 'scored_by', 'rank', 'popularity', 'members', 'favorites', 'premiered_year']
anime[numeric_columns] = scaler.fit_transform(anime[numeric_columns])
anime.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
anime['type'] = le.fit_transform(anime['type'])
anime['duration'] = le.fit_transform(anime['duration'])
anime['rating'] = le.fit_transform(anime['rating'])
anime['studio'] = le.fit_transform(anime['studio'])
anime['source'] = le.fit_transform(anime['source'])
anime['anime_duration'] = le.fit_transform(anime['anime_duration'])
anime.head()
```

```python
y = anime['title']
X = anime.drop('title', axis=1)
from sklearn.impute import KNNImputer
imputer = KNNImputer(n_neighbors=3)
X = pd.DataFrame(imputer.fit_transform(X), columns=X.columns, index=X.index)
X.head()
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)
X_pca = pd.DataFrame(X_pca, columns=['PC1', 'PC2'])
X_pca.head()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='PC1', y='PC2', data=X_pca)
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA of Anime Dataset')
plt.show()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
cosine_sim = cosine_similarity(X)
cosine_sim = pd.DataFrame(cosine_sim, index=y, columns=y)
cosine_sim.head()
```

```python
anime_id = 2001
anime_title = anime.loc[anime_id, 'title']
print(f"Anime Title: {anime_title}")
similar_anime = cosine_sim[anime_title].sort_values(ascending=False).drop(anime_title)
similar_anime.head(10)
```
